```python
#import dependency
import pandas as pd
```

```python
#read in csv exported from pgadmin
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
#total_votes count is equal to or greater than 20
vine_votes = vine_df[vine_df['total_votes']>=20]
vine_votes.head()
```

```python
#helpful_votes divided by total_votes is equal to or greater than 50%
vine_helpful_total_votes = vine_votes[vine_votes['helpful_votes']/vine_votes['total_votes']>=0.5]
vine_helpful_total_votes.head()
```

```python
#review was written as part of the Vine program (paid)
vine_paid = vine_helpful_total_votes[vine_helpful_total_votes['vine']=='Y']
vine_paid.head()
```

```python
#review was not part of the Vine program (unpaid)
vine_unpaid = vine_helpful_total_votes[vine_helpful_total_votes['vine']=='N']
vine_unpaid.head()
```

```python
#paid total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews
paid_count=len(vine_paid)
paid_fivestar_count = len(vine_paid[vine_paid['star_rating']==5.0])
paid_five_star_percentage = round(paid_fivestar_count/paid_count * 100,2)
#paid_five_star_percentage
```

```python
#unpaid total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews
unpaid_count=len(vine_unpaid)
unpaid_fivestar_count = len(vine_unpaid[vine_unpaid['star_rating']==5.0])
unpaid_five_star_percentage = round(unpaid_fivestar_count/unpaid_count * 100,2)
#unpaid_five_star_percentage
```

```python
print(f'There were a total of {paid_count} Vine reviews and {unpaid_count:,} non-Vine reviews.')
```

```python
print(f'There were a total of {paid_fivestar_count} Vine reviews that were 5-stars and {unpaid_fivestar_count:,} non-Vine Reviews that were 5-stars.')
```

```python
print(f'The percentage of Vine reviews that were 5-stars was {paid_five_star_percentage}%. The percentage of non-Vine reviews that were 5-stars was {unpaid_five_star_percentage}%.')
```
